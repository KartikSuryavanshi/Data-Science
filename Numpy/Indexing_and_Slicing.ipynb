```python
#Slicing returns a view, not a copy! Changes affect the original array.**This might seem counterintuitive since Python lists
#create copies when sliced.But in NumPy, slicing returns a view of the original array. Both the sliced array and the original
#array share the same data inmemory, so changes in the slice affect the original array.
```

```python
b = flat[3:7].copy() #this is not a slice because we are using copy method to get the copy.
```

```python
b
```

```python
b[0]= 23
```
