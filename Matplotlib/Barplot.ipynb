```python
import numpy as np
 
sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
 
x = np.arange(len(years))  # index positions
width = 0.25 #The value of width ranges from 0 to 1 (default 0.8); it can go above 1, but bars will start to overlap.

plt.bar(x - width, sachin, width=width, label="Sachin")
plt.bar(x, sehwag, width=width, label="Sehwag")
plt.bar(x + width, kohli, width=width, label="Kohli")
 
plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Run Comparison")
plt.xticks(x, years)  # Show actual year instead of 0,1,2,...Replaces index numbers with real labels
plt.legend()
plt.tight_layout() #Prevents labels from overlapping
```

```python
#Horizontal bar-chart : using barh ---Why Switch x and y?
#In bar()  x = categories, y = values
#In barh()  y = categories, x = values
#Horizontal bars help when category names are long or when you want to emphasize comparisons from left to right.


players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs_5yrs = [500+700+1100+1500+1800, 0+200+900+1400+1600, 0+0+500+800+1100, 300+600+800+1100+900]

plt.barh(players, runs_5yrs, color="skyblue")
plt.xlabel("Total Runs in First 5 Years")
plt.title("First 5-Year Performance of Indian Batsmen")
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

players = ["Sachin", "Sehwag", "Kohli"]
runs = [1500, 1200, 1800]

plt.bar(players, runs, color="#99eb95")

# Add labels on top of bars
for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center')
#Use ha='center' to center the text and add a small offset (+30 here) to avoid overlap with the bar.
plt.ylabel("Runs")
plt.title("Runs Scored by Players")
plt.tight_layout()
```

```python
#Create a side-by-side bar chart comparing runs of Sachin, Kohli, and Sehwag over 5 selected years.
#Then create a horizontal bar chart showing total runs scored in their debut 5 years.

# Selected years
years = [2000, 2001, 2002, 2003, 2004]

# Hypothetical yearly runs for each player
sachin = [500, 700, 1100, 1500, 1800]
kohli = [0, 200, 900, 1400, 1600]
sehwag = [0, 0, 500, 800, 1100]

# Side-by-side bar chart
x = np.arange(len(years))  # positions for each year
width = 0.25

plt.bar(x - width, sachin, width=width, label="Sachin", color="blue")
plt.bar(x, kohli, width=width, label="Kohli", color="orange")
plt.bar(x + width, sehwag, width=width, label="Sehwag", color="green")

plt.xlabel("Year")
plt.ylabel("Runs")
plt.title("Runs Comparison Over 5 Years")
plt.xticks(x, years)
plt.legend()
plt.tight_layout()

```

```python
# Horizontal bar chart of total debut 5-year runs
players = ["Sachin", "Kohli", "Sehwag"]
runs_5yrs = [sum(sachin), sum(kohli), sum(sehwag)]

x = np.arange(len(players))

plt.barh(x, runs_5yrs, color=["blue", "orange", "green"])
plt.yticks(x, players)
plt.xlabel("Total Runs")
plt.title("Total Runs in First 5 Years")
plt.tight_layout()
```
