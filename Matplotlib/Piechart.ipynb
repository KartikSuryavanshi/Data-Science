```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")  # Choose any style you like
```

```python
# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]

plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels)
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.pie(runs, labels=labels, colors=colors)
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'}
)
```

```python
explode = [0.1, 0, 0, 0]  # Only highlight Sachin's slice

plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True  # adds a 3D-like shadow
)
```

```python
plt.pie(
    runs,
    labels=labels,
    explode=explode,
    colors=colors,
    shadow=True,
    startangle=100
)
```

```python
plt.pie(
    runs,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140
)
plt.title("Career Run Share")
```

```python
# Too many categories---Crowded Pie Chart  Why to Avoid
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 'Swift', 'PHP']
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1]
 
plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage (Crowded Example)")
```

```python
#Create a pie chart showing the market share of mobile OS (Android, iOS, others). 
#Then recreate the same using a horizontal bar chart and observe which is easier to understand.


#Pie chart
market_share_of_mobile = ['Android', 'IOS', 'Others']
share = [30, 15,5]
 
plt.pie(usage, labels=market_share_of_mobile, autopct='%1.1f%%', startangle=90)
plt.title("Market share of mobile OS (Android, iOS, others)")
```

```python
#Horizontal bar chart

plt.barh(market_share_of_mobile,share, color="red")
plt.xlabel("Market share of mobile OS")
plt.title("Market share of mobile OS (Android, iOS, others)")
plt.tight_layout()
```

```python
# Therefore pie chart is more easier to understand..
```
